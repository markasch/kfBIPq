## Exercise 1
How do the dynamics of the covariance behave for the one dimensional Kalman filter. Consider $|\lambda|<1$ and $|\lambda|>1$ as well as $ \sigma^2=0$ and $\sigma^2>0$

```python
def kalmanfilter_onedim(c_0,gamma,sigma,lambd,N):
    # Inputs:
    # c_0: starting covariance 
    # gamma: covariance of the noise in the data
    # sigma: covariance of the noise in the forward model
    # lambd: model parameter
    # N: amount of iterations
    
    # Outputs:
    # c: N+1-vector with the computed covariances
    # c_lim_plus, c_lim_minus: limiting covariances
    
    c=np.zeros(N+1)
    c[0]=c_0
    gamma_sq=gamma**2
    sigma_sq=sigma**2
    lambd_sq=lambd**2
    
    c_lim_plus=-(gamma_sq+sigma_sq-gamma_sq*lambd_sq)+np.sqrt((gamma_sq+sigma_sq-gamma_sq*lambd_sq)**2+4*lambd_sq*gamma_sq*sigma_sq)
    c_lim_plus=c_lim_plus/(2*lambd_sq)
    
    c_lim_min=-(gamma_sq+sigma_sq-gamma_sq*lambd_sq)-np.sqrt((gamma_sq+sigma_sq-gamma_sq*lambd_sq)**2+4*lambd_sq*gamma_sq*sigma_sq)
    c_lim_min=c_lim_min/(2*lambd_sq)
    
    for j in range(1,N+1):
        c[j]=(gamma_sq*(lambd_sq*c[j-1]+sigma_sq))/(gamma_sq+lambd_sq*c[j-1]+sigma_sq)
    
    return c, c_lim_plus, c_lim_min
```

```python
# Set parameters for lambda<1
c_0=1
gamma=1
sigma=0
lambd=0.8
N=50

c_l1,c_p_l1,c_m_l1=kalmanfilter_onedim(c_0,gamma,sigma,lambd,N)

# Set parameters for lambda > 1
c_0=1
gamma=1
sigma=0
lambd=1.2
N=50

c_l2,c_p_l2,c_m_l2=kalmanfilter_onedim(c_0,gamma,sigma,lambd,N)

# Set parameters for lambda = 1
c_0=1
gamma=1
sigma=0
lambd=1
N=50

c_l3,c_p_l3,c_m_l3=kalmanfilter_onedim(c_0,gamma,sigma,lambd,N)

# Set parameters for sigma > 0
c_0=1
gamma=1
sigma=0.5
lambd=1.2
N=50

c_l4,c_p_l4,c_m_l4=kalmanfilter_onedim(c_0,gamma,sigma,lambd,N)

#Define parameters for plots:
iterations=list(range(1,N+2))
c_lim_p_l1=np.ones(N+1)*c_p_l1
c_lim_m_l1=np.ones(N+1)*c_m_l1

c_lim_p_l2=np.ones(N+1)*c_p_l2
c_lim_m_l2=np.ones(N+1)*c_m_l2

c_lim_p_l3=np.ones(N+1)*c_p_l3
c_lim_m_l3=np.ones(N+1)*c_m_l3

c_lim_p_l4=np.ones(N+1)*c_p_l4
c_lim_m_l4=np.ones(N+1)*c_m_l4

#Plot results

fig,axs=plt.subplots(2,2)
axs[0,0].loglog(iterations,c_l1-c_lim_p_l1,
         iterations,c_l1,"or")
axs[0,0].set_title(r'$\gamma=1, \sigma=0, \lambda=0.8$')
axs[0,1].loglog(iterations,c_l2-c_lim_p_l2,
         iterations,c_l2-c_lim_p_l2,"or")
axs[0,1].set_title('$\gamma=1, \sigma=0, \lambda=1.2$')
axs[1,0].loglog(iterations,c_l3-c_lim_m_l3,
         iterations,c_l3-c_lim_m_l3,"or")
axs[1,0].set_title('$\gamma=1, \sigma=0, \lambda=1$')
axs[1,1].loglog(iterations,c_l4-c_lim_p_l4,
         iterations,c_l4-c_lim_p_l4,"or")
axs[1,1].set_title('$\gamma=1, \sigma=0.5, \lambda=1.2$')

plt.tight_layout(pad=1.08,h_pad=None,w_pad=None,rect=None)
plt.rcParams['figure.figsize']=[15,5]
```

## Implement the one dimensional EnKF algorithm for arbitrary operator $\Psi$ and a linear operator $H$.

```python
def enkf_one_dim(m_0,C_0,N,Psi,H,sigma,gamma,y):
    # Inputs:
    # m_0, C_0: mean value and covariance of inital ensemble
    # N: amount of iterations
    # Psi, H: one-dimensional forward operators of the model
    # gamma: covariance of the noise in the data
    # sigma: covariance of the noise in the forward model
    # y: observed data 

    
    # Outputs:
    # U: JxN matrix with the computed particles for each iteration
    # m: vector of length N with the mean value of the particles
    # c: vector of length N with the covariance of the particles
    
    m=np.zeros(N)
    c=np.zeros(N)
    U=np.zeros((J,N))
    

    #Construct initial ensemble and estimator
    u_0=np.random.normal(m_0,C_0,J)
    U[:,0]=u_0
    m[0]=np.mean(U[:,0])
    c[0]=(U[:,0]-m[0])@(U[:,0]-m[0]).T/(J-1)

    
    for n in range(1,N):
        
        U_pred=Psi(U[:,n-1])+np.random.normal(0,sigma,J)
        mhat=np.mean(U_pred)
        chat=(U_pred-mhat)@(U_pred-mhat).T/(J-1)
        
        K=chat*H/(H*chat*H+gamma**2)
        d=y[n]*np.ones(J)+np.random.normal(0,gamma,J)-H*U_pred
        U[:,n]=U_pred+K*d
        m[n]=np.mean(U[:,n])
        c[n]=(U[:,n]-m[n])@(U[:,n]-m[n]).T/(J-1)
        
    
    
    return U,m,c
```

### Implement the one dimensional operators $\Psi$ given by example 2.3 and 2.5

```python
def forward_sin(z,alpha):
    return alpha*np.sin(z)

def forward_exp(z,L,h):
    return np.exp(L*h)*z

def forward_log(z,k,r,h):
    return k/(1+np.exp(-r*k*h)*(k/z-1))
```

### Now construct a sequence of data given a random starting value v_0 using both forward operators $\Psi$ from example 2.3 and 2.5 and apply the EnKF on this data.

We start with the sin function

```python
#Set Parameters
J=10
gamma=1
alpha=2.5
sigma=1e-3
H=1
m_0=0
C_0=9e-2
m_true=0
c_true=9e-1
N=1000


v=np.zeros(N)
y=np.zeros(N)

#Construct observation given the true vector
v[0]=np.random.normal(m_true,c_true)
y[0]=H*v[0]+np.random.normal(0,gamma)

for n in range(1,N):
        #Update true Parameter and data
        v[n]=forward_sin(v[n-1],alpha)+np.random.normal(0,sigma)
        y[n]=H*v[n]+np.random.normal(0,gamma)
        
# Call EnKF with the sin function as forward operator
# First we need to define a partial function that takes the pre used parameters in \Psi

partial_sin=functools.partial(forward_sin,alpha=2.5)
    
        
U,m,c=enkf_one_dim(m_0,C_0,N,partial_sin,H,sigma,gamma,y)
```

### Plot results 

```python
it=60
iterations=list(range(1,it+1))
plt.xlabel('Iteration Number')
plt.ylabel('Error')
plt.plot(iterations,v[0:it],"r",label='$v$')
plt.plot(iterations,m[0:it],"g",label='$m$')
plt.plot(iterations,m[0:it]+2*np.sqrt(c[0:it]),"b--",label='$m\pm2*\sqrt{c}$')
plt.plot(iterations,m[0:it]-2*np.sqrt(c[0:it]),"b--")
plt.legend(loc="lower right")
```

```python
it=500
iterations=list(range(1,it+1))
plt.xlabel('Iteration Number n')
plt.ylabel('Covariance')
plt.plot(iterations,c[0:it],"r", label='$c_n$')
plt.plot(iterations,np.divide(np.cumsum(c[0:it]),iterations)[0:it],"g",label='$N^{-1}\sum_k^N c_n$')
plt.legend(loc="upper right")
```

```python
it=50
iterations=list(range(1,it+1))
plt.xlabel('Iteration Number')
plt.ylabel('Error')
plt.plot(iterations,(v-m)[0:it]**2,"r", label='$(v_n-m_n)^2$')
plt.plot(iterations,np.divide((v-m)[0:it]**2,iterations),"g",label='$N^{-1}\sum_k^N (v_n-m_n)^2$')
plt.legend(loc="upper right")
```

## Implement the one dimensional SIRS particle filter for arbitrary $\Psi$ and linear H

```python
def sirs_one_dim(m_0,C_0,N,Psi,H,sigma,gamma,y):
    # Inputs:
    # m_0, C_0: mean value and covariance of inital ensemble
    # N: amount of iterations
    # Psi, H: one-dimensional forward operators of the model
    # gamma: covariance of the noise in the data
    # sigma: covariance of the noise in the forward model
    # y: observed data 

    
    # Outputs:
    # U: JxN matrix with the computed particles for each iteration
    # m: vector of length N with the mean value of the particles
    # c: vector of length N with the covariance of the particles
    
    m=np.zeros(N)
    c=np.zeros(N)
    U=np.zeros((J,N))
    

    #Construct initial ensemble and estimator
    u_0=np.random.normal(m_0,C_0,J)
    U[:,0]=u_0
    m[0]=np.mean(U[:,0])
    c[0]=(U[:,0]-m[0])@(U[:,0]-m[0]).T/(J-1)

    
    for n in range(1,N):
        
        U_pred=Psi(U[:,n-1])+np.random.normal(0,sigma,J)
        d=y[n]*np.ones(J)-H*U_pred
        what=np.exp(-0.5*((1/gamma**2)*d**2))
        w=what/np.sum(what)
        ws=np.cumsum(w)
        
        for j in range(J):
            ix=np.searchsorted(ws,np.random.uniform(0,1),side='right')
            U[j,n]=U_pred[ix]
        
        m[n]=np.mean(U[:,n])
        c[n]=(U[:,n]-m[n])@(U[:,n]-m[n]).T/(J-1)
        
    
    
    return U,m,c
```

```python
#Set Parameters
J=1000
gamma=1
alpha=2.5
sigma=1e-3
H=1
m_0=0
C_0=9e-2
m_true=0
c_true=9e-1
N=1000


v=np.zeros(N)
y=np.zeros(N)

#Construct observation given the true vector
v[0]=np.random.normal(m_true,c_true)
y[0]=H*v[0]+np.random.normal(0,gamma)

for n in range(1,N):
        #Update true Parameter and data
        v[n]=forward_sin(v[n-1],alpha)+np.random.normal(0,sigma)
        y[n]=H*v[n]+np.random.normal(0,gamma)
        
# Call SIRS with the sin function as forward operator
# First we need to define a partial function that takes the pre used parameters in \Psi

partial_sin=functools.partial(forward_sin,alpha=2.5)
    
        
U,m,c=sirs_one_dim(m_0,C_0,N,partial_sin,H,sigma,gamma,y)
```

## Implement the one-dimensional EKI for a linear forward operator g

```python
def eki_one_dim_lin(m_0,C_0,N,G,sigma,gamma,y,delt,h):
    # Inputs:
    # m_0, C_0: mean value and covariance of inital ensemble
    # N: amount of iterations
    # G: one-dimensional forward operators of the model
    # gamma: covariance of the noise in the data
    # sigma: covariance of the noise in the forward model
    # y: observed data 
    # h: discretization step

    
    # Outputs:
    # U: JxN matrix with the computed particles for each iteration
    # m: vector of length N with the mean value of the particles
    # c: vector of length N with the covariance of the particles
    
    m=np.zeros(N)
    c=np.zeros(N)
    U=np.zeros((J,N))
    

    #Construct initial ensemble and estimator
    u_0=np.random.normal(m_0,C_0,J)
    U[:,0]=u_0
    m[0]=np.mean(U[:,0])
    c[0]=(U[:,0]-m[0])@(U[:,0]-m[0]).T/(J-1)
    
    for n in range(1,N):
        
        # Last iterate under forward operator:
        G_u=G*U[:,n-1]
        Ghat=np.mean(G_u)
        U[:,n]=U[:,n-1]+h*(c[n-1]+delt)*G*(1/gamma)*((y-G_u))
        
        
        m[n]=np.mean(U[:,n])
        c[n]=(U[:,n]-m[n])@(U[:,n]-m[n]).T/(J-1)
        
    
    
    return U,m,c
```

```python
#Set Parameters
J=10
gamma=1
sigma=1e-3
m_0=0
C_0=9e-1
m_true=0
c_true=C_0
G=1.5
N=10000
h=1/100
delt=1

# Construct data under true parameter
u_true=np.random.normal(m_true,c_true)
y=G*u_true

U,m,c=eki_one_dim_lin(m_0,C_0,N,G,sigma,gamma,y,delt,h)
```

```python
it=N
iterations=list(range(1,(it+1)))
plt.xlabel('Iterations number n')
plt.ylabel('Error')
plt.loglog(iterations,(u_true*np.ones(N)-m)**2,"r",label='$u^\dagger-m_n$')
plt.legend(loc="upper right")
```

```python
it=300
iterations=list(range(1,it+1))
plt.xlabel('Iteration number n')
plt.ylabel('Covariance')
plt.plot(iterations,c[0:it],"r", label='$c_n$')
plt.plot(iterations,np.divide(np.cumsum(c[0:it]),iterations)[0:it],"g",label='$N^{-1}\sum_k^N c_n$')
plt.legend(loc="upper right")
```

## Implement the one dimensional EKI for an arbitrary forward operator $G$.

```python
def eki_one_dim(m_0,C_0,N,G,sigma,gamma,y,h):
    # Inputs:
    # m_0, C_0: mean value and covariance of inital ensemble
    # N: amount of iterations
    # G: one-dimensional forward operators of the model
    # gamma: covariance of the noise in the data
    # sigma: covariance of the noise in the forward model
    # y: observed data
    # h: discretization step

    
    # Outputs:
    # U: JxN matrix with the computed particles for each iteration
    # m: vector of length N with the mean value of the particles
    # c: vector of length N with the covariance of the particles
    
    m=np.zeros(N)
    c=np.zeros(N)
    U=np.zeros((J,N))
    

    #Construct initial ensemble and estimator
    u_0=np.random.normal(m_0,C_0,J)
    U[:,0]=u_0
    m[0]=np.mean(U[:,0])
    c[0]=(U[:,0]-m[0])@(U[:,0]-m[0]).T/(J-1)
    
    for n in range(1,N):
        
        # Last iterate under forward operator:
        G_u=G(U[:,n-1])
        uhat=np.mean(U[:,n-1])
        Ghat=np.mean(G_u)
        
        cov_uG=(U[:,n-1]-uhat)@(G_u-Ghat).T/(J-1)
        cov_GG=(G_u-Ghat)@(G_u-Ghat).T/(J-1)
        
        U[:,n]=U[:,n-1]+cov_uG*h/(h*cov_GG+gamma)*(y-G_u)        
        
        m[n]=np.mean(U[:,n])
        c[n]=(U[:,n]-m[n])@(U[:,n]-m[n]).T/(J-1)
        
    
    
    return U,m,c
```

```python
#Set Parameters
J=10
r=10
k=10
gamma=1
sigma=1e-3
m_0=0
C_0=9e-2
m_true=0
c_true=C_0
N=10000
h=1/100

# Construct data under true parameter
u_true=np.random.normal(m_true,c_true)
y=forward_log(u_true,k,r,h)

# Use partial function
partial_log=functools.partial(forward_log,k=10,r=10,h=1/100)

U,m,c=eki_one_dim(m_0,C_0,N,partial_log,sigma,gamma,y,h)
```

```python
it=N
iterations=list(range(1,it+1))
plt.xlabel('Iterations number n')
plt.ylabel('Error')
plt.plot(iterations,(u_true*np.ones(N)-m)**2,"r",label='$u^\dagger-m_n$')
plt.legend(loc="upper right")
```

```python
it=300
iterations=list(range(1,it+1))
plt.xlabel('Iteration number n')
plt.ylabel('Covariance')
plt.plot(iterations,c[0:it],"r", label='$c_n$')
plt.plot(iterations,np.divide(np.cumsum(c[0:it]),iterations)[0:it],"g",label='$N^{-1}\sum_k^N c_n$')
plt.legend(loc="upper right")
```
